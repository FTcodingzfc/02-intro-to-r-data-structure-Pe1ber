```r
#使用data.table包
library(data.table)
library(stringr)
```

1.将posts和replys数据集导入R中，命名为r.posts和r.replys，保留所有字段名，所有字段空⽩转化为NA

```r
r_replys <- fread("replys.csv", na.strings = "")
r_posts <- fread("posts.csv", na.strings = "")
```

删除所有postnums字段都为0的⾏，查看post_id的数量

```r
r_posts <- r_posts[postnums!=0]
post_id <- r_posts[,post_id]
num_post_id <-length(post_id)
print(num_post_id)
```

统计字段guba_name为**“螺纹钢吧”**帖⼦的数量，与postnums进⾏对比，哪⼀个更⼤

```r
num_luowengang <- nrow(r_posts[guba_name=="螺纹钢吧"])
print(num_luowengang)
postnums_value <- r_posts[guba_name == "螺纹钢吧", postnums][1]
print(postnums_value)
if (num_luowengang>postnums_value){print("帖子数量更大")} else if(num_luowengang<postnums_value){print("postnums更大")} else {print("相等")}
```

按create_time进⾏升序排列，找出每个吧中最早创建的帖⼦，将所有这些帖⼦的信息导出⽣成⼀个表,命名为earl.posts

```r
earl.posts <- r_posts[order(create_time), .SD[1],by=guba_name]
print(earl.posts)
fwrite(earl.posts, "earl_posts.csv")
```

2.按照post_id和reply_time进⾏分组升序排列，导出最晚的⼀条回
复，⽣成⼀个表，名为late.replys


```r
late_replys<-r_replys[order(post_id,reply_time),.SD[.N],by=post_id]
print(late_replys)
fwrite(late_replys,"late.replys.csv")
```

删除所有reply_content中带有’<>’的部分，例如<imgsrc=""http://gbres.dfcfw.com/face/emot/emot14.png""title=""⽜"">，将此类内容删除

```r
late_replys$reply_content<- str_remove_all(late_replys$reply_content, "<[^>]*>")
print(late_replys)
fwrite(late_replys,"late_replys.csv")
```

4.上述操作完成之后，对已经处理好的r.posts和r.replys，以post_id这个字段进⾏合并，⽣成⼀个
表，命名为int.data。要求每⼀条reply都对应⼀条帖⼦的信息，也就是将帖⼦中的reply全部展开，
删除在两表中⽆法匹配的条⽬

```r

int.data<-r_posts[r_replys,on="post_id",nomatch=0]
print(int.data)
fwrite(int.data,"int.data.csv")
```

5.提取int.data中create_time的年⽉⽇部分，⽣成⼀个新变量create_date，以"%Y-%m-%d"的Date格
式储存；将create_time改为POSXct的标准时间格式："%Y-%m-%d %H:%M:%OS"。

```r
int.data[, create_time := as.POSIXct(create_time, format = "%Y-%m-%d %H:%M:%OS")]
int.data[, create_date := as.Date(create_time,format="%Y-%m-%d")]
print(int.data$create_date)
```

提取int.data中create_date的最⼤值和最⼩值，提取所有guba_name，对每⼀个guba_name都⽣成⼀
个以⽇为单位的时间序列向量，区间为create_date的最⼤值和最⼩值。并最后输出⼀个表，命名为tim

```r
min_date <- min(int.data$create_date)
max_date <- max(int.data$create_date)
all_guba_names <- unique(int.data$guba_name)
full_date_seq <- seq(from = min_date, to = max_date, by = "day")
tim <- CJ(guba_name = all_guba_names, create_date = full_date_seq)
print(tim)
```

将tim与int.data以guba_name和create_date进⾏合并，⽣成⼀个新表，命名为int.data.ts，将所
有为NA的readnum和replynum观测填充为0

```r
int.data.ts<-int.data[tim,on=c("guba_name","create_date")]
int.data.ts[is.na(readnum), readnum := 0]
int.data.ts[is.na(replynum), replynum := 0]
print(int.data.ts)
```

6.在r.posts中，提取出replynum最多的⼗条帖⼦和⼗个吧，⽣成两张表为mx.post和mx.ba，表⾥变量
分别包括帖⼦id、吧名、帖⼦阅读数和帖⼦回复数，吧名、吧总体阅读数和吧总体回复数

```r
mx.post <- r_posts[order(-replynum)][1:10, .(post_id,guba_name,readnum,replynum)]
fwrite(mx.post, "mx_post.csv")
mx.ba <- r_posts[, .(total_readnum = sum(readnum, na.rm = TRUE), total_replynum = sum(replynum, na.rm = TRUE)),by = guba_name][order(-total_replynum)][1:10]
fwrite(mx.ba, "mx_ba.csv")
```

在r.posts中，调⽤任何相关的描述性统计包，对帖⼦和吧做描述性统计，要求⾄少导出replynum和
readnum的最⼤值、最⼩值、均值、中位数和⽅差，并导成⼀张表名为stst.int

```r
stst.int <- r_posts[, .( 
  replynum_max = max(replynum, na.rm = TRUE),
  replynum_min = min(replynum, na.rm = TRUE),
  replynum_mean = mean(replynum, na.rm = TRUE),
  replynum_median = median(replynum, na.rm = TRUE),
  replynum_var = var(replynum, na.rm = TRUE),
  readnum_max = max(readnum, na.rm = TRUE),
  readnum_min = min(readnum, na.rm = TRUE),
  readnum_mean = mean(readnum, na.rm = TRUE),
  readnum_median = median(readnum, na.rm = TRUE),
  readnum_var = var(readnum, na.rm = TRUE)
)]
fwrite(stst.int, "stst_int.csv")
```
